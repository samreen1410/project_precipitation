# Reading Data

```python
#importing libraries
import pandas as pd
import numpy as np
import glob
import os
```

```python
#monthly flow rate data from 1925-2924
flow_monthly = pd.read_csv("monthly_flow_rate_fraser.csv", skiprows = 1)
flow_monthly.head(15)
```

```python
#monthly temperature data from 1925 - 2007
temp_monthly_1925_2007 = pd.read_csv("monthly_1925-2007.csv")
temp_monthly_1925_2007.head()
```

```python
# Get all files that start with 'daily_20'
daily_files = glob.glob("daily_20*.csv")
daily_files
```

```python
#reading and merging all daily temperature data files from 2008 - 2024
daily_list = []
for file in daily_files:
    df = pd.read_csv(file)
    daily_list.append(df)
temp_daily_2008_2024 = pd.concat(daily_list, ignore_index=True)
temp_daily_2008_2024.head()
```

```python
print("Flow monthly shape:", flow_monthly.shape)
print("Monthly temp shape:", temp_monthly_1925_2007.shape)
print("Daily temp shape:", temp_daily_2008_2024.shape)
```

# Cleaning flow data

```python
#dropping unwanted columns from flow rate data
flow_clean = flow_monthly.drop(columns=[' ID', 'TYPE', 'Mean'])
flow_clean = flow_clean.loc[:, ~flow_clean.columns.str.contains('SYM')]
flow_clean = flow_clean[(flow_clean['Year'] >= 1925) & (flow_clean['Year'] <= 2024)]
flow_clean.head()
```

```python
#changing the format of the dataset to match others
flow = flow_clean.melt(
    id_vars=['PARAM', 'Year'],
    var_name='Month',
    value_name='Discharge'
)
flow.head()
```

```python
#converting month names to numbers as in other datasets
month_dict = {
    'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4,
    'May':5, 'Jun':6, 'Jul':7, 'Aug':8,
    'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12
}
flow['Month'] = flow['Month'].map(month_dict)
```

# Cleaning temp data

```python
temp_daily_2008_2024.columns
```

```python
temp_monthly_1925_2007.columns
```

```python
# taking necessary columns and making sure numeric values
temp_monthly_1925_2007 = temp_monthly_1925_2007[
    ['Year','Month','Total Precip (mm)','Mean Temp (°C)']
].copy()
temp_monthly_1925_2007['Mean Temp (°C)'] = pd.to_numeric(
    temp_monthly_1925_2007['Mean Temp (°C)'],
    errors='coerce'
)
temp_monthly_1925_2007['Total Precip (mm)'] = pd.to_numeric(
    temp_monthly_1925_2007['Total Precip (mm)'],
    errors='coerce'
)
```

```python
# making sure numeric values
daily_clean = temp_daily_2008_2024.copy()
daily_clean['Mean Temp (°C)'] = pd.to_numeric(
    daily_clean['Mean Temp (°C)'],
    errors='coerce'
)
daily_clean['Total Precip (mm)'] = pd.to_numeric(
    daily_clean['Total Precip (mm)'],
    errors='coerce'
)
```

```python
# get monthly temp avergaes from daily data
temp_monthly_2008_2024 = (
    daily_clean
    .groupby(['Year','Month'])['Mean Temp (°C)']
    .mean()
    .reset_index()
)
```

```python
# get monthly prec avergaes from daily data
prec_monthly_2008_2024 = (
    daily_clean
    .groupby(['Year','Month'])['Total Precip (mm)']
    .sum()
    .reset_index()
)
```

```python
#renaming columns 
monthly_1925_2007 = temp_monthly_1925_2007[
    ['Year','Month','Mean Temp (°C)','Total Precip (mm)']
].rename(columns={
    'Mean Temp (°C)': 'Temp',
    'Total Precip (mm)': 'Prec'
})
temp_2008_2024 = temp_monthly_2008_2024.rename(
    columns={'Mean Temp (°C)': 'Temp'}
)
prec_2008_2024 = prec_monthly_2008_2024.rename(
    columns={'Total Precip (mm)': 'Prec'}
)
```

```python
#merging datasets to get ne dataset with monthly temp and prec from 
#1925-2024
climate = pd.merge(
    temp_2008_2024,
    prec_2008_2024,
    on=['Year','Month'],
    how='inner'
)
climate = pd.concat(
    [monthly_1925_2007, climate],
    ignore_index=True
)
#climate = climate.sort_values(
 #   ['Year','Month']
#).reset_index(drop=True)
print(climate.head())
print("Shape:", climate.shape)
```

```python
#thus our final cleaned 2 datasets:
print(flow.head())
print(climate.head())
```

# Missing data analysis

```python
def missing_data(df, name):
    print(f" Missing Data {name}")
    total_cells = df.size
    missing_cells = df.isna().sum().sum()
    missing_percent = (missing_cells / total_cells) * 100
    print(f"Total cells: {total_cells}")
    print(f"Missing cells: {missing_cells}")
    print(f"Missing percentage: {missing_percent:.2f}%")
```

```python
missing_data(flow, "Flow Dataset")
missing_data(climate, "Climate Dataset")
```

the missing percentage is not that significant. we can handle this. lets check which columns have the most missing values

```python
def missing_by_column(df, name):
    print(f"Missing Values by Column:{name}")
    col_missing = df.isna().sum().sort_values(ascending=False)
    col_percent = df.isna().mean().sort_values(ascending=False) * 100
    result = pd.DataFrame({
        "Missing Count": col_missing,
        "Missing Percentage": col_percent
    })
    print(result)
missing_by_column(flow, "Flow Dataset")
missing_by_column(climate, "Climate Dataset")
```

the missing data for temperature is alot.

```python
#lets try to drop it
climate = climate.dropna().reset_index(drop=True)
flow = flow.dropna().reset_index(drop=True)
```

```python
print("Flow dataset shape:", flow.shape)
print("Climate dataset shape:", climate.shape)

print("\nMissing values remaining:")
print(flow.isna().sum())
print(climate.isna().sum())
```

```python
flow_years = flow['Year'].unique()
climate_years = climate['Year'].unique()

print("Flow dataset years:", sorted(flow_years))
print("Climate dataset years:", sorted(climate_years))
```

**we see that for flow data, we have years from 1925 - 2024 but for climate data after dropping na we have years from 1958-2024**

# Initial Patterns

```python
import matplotlib.pyplot as plt
```

```python
#annual trends
flow_annual = flow.groupby('Year')['Discharge'].mean().reset_index()
climate_annual = climate.groupby('Year')[['Temp','Prec']].mean().reset_index()
```

```python
#annual discharge trend
plt.figure(figsize=(8,5))
plt.plot(flow_annual['Year'], flow_annual['Discharge'])
plt.xlabel("Year")
plt.ylabel("Mean Annual Discharge")
plt.title("Annual Discharge Trend")
plt.show()
```

```python
#annnual temperature trend
plt.figure(figsize=(8,5))
plt.plot(climate_annual['Year'], climate_annual['Temp'])
plt.xlabel("Year")
plt.ylabel("Mean Annual Temperature")
plt.title("Annual Temperature Trend")
plt.show()
```

```python
#annual precipitation trend
plt.figure(figsize=(8,5))
plt.plot(climate_annual['Year'], climate_annual['Prec'])
plt.xlabel("Year")
plt.ylabel("Mean Annual Precipitation (mm)")
plt.title("Annual Precipitation Trend")
plt.show()
```

```python
#plot for seeing relationship between temp and discharge
#merging first to keep x and y of the same size
df_plot = pd.merge(
    climate_annual,
    flow_annual,
    on='Year',
    how='inner'
)
plt.figure(figsize=(6,5))
plt.scatter(
    df_plot['Temp'],
    df_plot['Discharge'],
    s=20)
plt.xlabel("Annual Mean Temperature")
plt.ylabel("Annual Mean Discharge")
plt.title("Temperature vs Discharge Relationship")
plt.show()
```
